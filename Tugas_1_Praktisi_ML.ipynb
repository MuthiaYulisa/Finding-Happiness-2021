**Exploratory Data Analysis (EDA)**

```python
features = ['Ladder score', 'Standard error of ladder score', 'upperwhisker', 'lowerwhisker', 'Logged GDP per capita',
            'Social support', 'Healthy life expectancy', 'Freedom to make life choices',	'Generosity', 'Perceptions of corruption',
            'Ladder score in Dystopia', 'Explained by: Log GDP per capita', 'Explained by: Social support',	'Explained by: Healthy life expectancy',
            'Explained by: Freedom to make life choices', 'Explained by: Generosity', 'Explained by: Perceptions of corruption',	'Dystopia + residual']
```

```python
numeric_data = df[features]
```

```python
plt.figure(figsize=(14, 6))
sns.heatmap(numeric_data.corr(method='pearson'), annot=True, cmap='inferno')
plt.title('Korelasi antara Fitur Indikator Kebahagiaan')
plt.show()
```

Dari heatmap tersebut kita sudah mendapatkan korelasi untuk melakukan analisis, korelasi yang kuat untuk mencari target variable score diantaranya ada GDP per capita, Social Support dan Healthy life expectancy. tapi kita juga dapat melakukan analisa untuk fitur yang lain seperti Freedom to make life choices.

*Analisis Hubungan Pendapatan per kapita dengan Tingkat Kebahagiaan*

```python
plt.figure(figsize=(20,6))
sns.lineplot(x=df["Logged GDP per capita"], y=df["Ladder score"], marker='D', color='red', linestyle='-')

plt.title("Komparasi GDP per capita vs Score", size=15)
plt.xlabel("GDP per capita")
plt.ylabel("Score")

plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

Berdasarkan Visualisasi dari Dataset tersebut, kita melihat korelasi yang Fluktuatif, tetapi semakin tinggi skor kebahagiannya, standar GDP per capitanya pun juga ikut naik.

*Analisis Hubungan Dukungan Sosial terhadap Tingkat Kebahagiaan*

```python
plt.figure(figsize=(20,6))
sns.lineplot(x=df["Social support"], y=df["Ladder score"], marker='D', color='purple', linestyle='-')

plt.title("Komparasi Social support vs Score", size=15)
plt.xlabel("Social Support")
plt.ylabel("Score")

plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

dari visualisasi dataset yang ada, hasil analisisnya adalah ternyata pengaruh dukungan sosial juga dapat mempengaruhi tingkat kebahagiaan seseorang. Karena, secara logika ketika seseorang mendapatkan support sosial mereka akan menjalani hidup dengan senang dan itu akan membuat orang itu semakin bahagia. Tetapi, dengan melihat grafik tersebut, kita masih belum bisa memastikan karena data yang ada masih fluktuatif.


*Analisis Hubungan Kebebasan dengan Tingkat Kebahagiaan*

```python
plt.figure(figsize=(20, 6))
sns.scatterplot(data=df, x='Freedom to make life choices', y='Ladder score', alpha=1)

plt.title('Komparasi Freedom to make life choices vs Score', size=15)
plt.xlabel('Freedom to make life choices')
plt.ylabel('Score')

plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```


Berdasarkan Visualisasi Dataset kita dapat bahwa masih hasilnya fluktuatif, walaupun begitu terlihat juga kalau semakin tinggi tingkat kebebasan seseorang semakin tinggi juga tingkat kebahagiannya.

*Analisis Hubungan Harapan Hidup Sehat dengan Tingkat Kebahagiaan*

```python
plt.figure(figsize=(20, 6))
sns.scatterplot(data=df, x='Healthy life expectancy', y='Ladder score', alpha=1)

plt.title('Komparasi Healthy life expectancy vs Score', size=15)
plt.xlabel('Healthy life expectancy')
plt.ylabel('Score')

plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

Healthy life expectancy atau harapan hidup sehat dapat mempengaruhi tingkat kebahagiaan seseorang. Menurut beberapa penelitian, ada hubungan positif antara harapan hidup sehat dan tingkat kebahagiaan.
